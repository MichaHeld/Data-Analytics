```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd

# CSV-Datei von der Festplatte laden
file_path = "data/Obesity_Dataset.xlsx"
df = pd.read_excel(file_path)

# Boxplot erstellen
plt.figure(figsize=(10, 6))
sns.boxplot(x='Number_of_Main_Meals_Daily', y='Age', data=df, palette="Set2")

# Titel und Achsenbeschriftungen
plt.title('Alter je nach Menge von Hauptmahlzeiten')
plt.xlabel('Menge von Hauptmahlzeiten')
plt.ylabel('Alter')

# Y-Achse auf 5er Schritte beschränken
plt.yticks(range(18, 55, 5))  # Altersbereich von 18 bis 54 in 5er Schritten

# X-Achse sicherstellen, dass sie nur die Kategorien 1, 2, 3 anzeigt
plt.xticks([0, 1, 2], labels=['1-2 Mal/Tag', '3 Mal/Tag', '3+ Mal/Tag'])  # Optionale Labels

# Hilfslinien für bessere Lesbarkeit
plt.grid(axis='y', linestyle='--', alpha=0.7)

# Diagramm anzeigen
plt.show()

```
