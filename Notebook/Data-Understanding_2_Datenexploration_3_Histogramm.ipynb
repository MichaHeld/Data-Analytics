```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd

# CSV-Datei von der Festplatte laden
file_path = "data/Obesity_Dataset.xlsx"
df = pd.read_excel(file_path)

# Manuelle Definition der Bins (6-Jahres-Intervalle)
bins = range(18, 55, 6)  # Start bei 18, Ende bei 54, Schrittweite von 6

# Visualisierung
plt.figure(figsize=(10, 6))
sns.histplot(data=df, x='Age', bins=bins, kde=True, color='skyblue', line_kws={'lw': 2})  # Nutzt die Bins mit KDE
plt.title('Verteilung des Alters (6-Jahres-Intervalle) mit KDE')
plt.xlabel('Alter (6-Jahres-Intervalle)')
plt.ylabel('Häufigkeit / Dichte')
plt.xticks(bins)  # Anpassung der x-Achsen-Ticks an die Bins
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()

```
