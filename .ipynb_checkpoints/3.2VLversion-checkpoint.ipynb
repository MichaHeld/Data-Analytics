```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd


# CSV-Datei von der Festplatte laden
file_path = r"C:\Users\Michahaha\Documents\Data Analytics\projektdaten\data\Obesity_Dataset.xlsx"

df = pd.read_excel(file_path)

df_bmi = pd.DataFrame({
   'Class': [ 1, 2, 3, 4],
    'BMI': ['Unter 18.5', '18.5-24.9', '25-29.9', '30 höher']
})

# Zusammenführen der beiden DataFrames basierend auf der "Sex"-Spalte
df_merged = pd.merge(df, df_bmi, on='Class', how='left')

# Aufgaebe 1: Anzeigen der ersten Zeilen des zusammengeführten DataFrames
#print(df_merged.head())

#Aufgabe 3: 
# Erstellen eines Balkendiagramms zur Visualisierung der Häufigkeitsverteilung
# Durchschnittliches Alter pro BMI-Kategorie berechnen
age_per_bmi = df_merged.groupby('BMI')['Age'].mean()

# Balkendiagramm erstellen
plt.figure(figsize=(10, 6))
sns.barplot(x=age_per_bmi.index, y=age_per_bmi.values, palette=['#4CAF50', '#2196F3', '#FFC107', '#F44336'])
plt.xlabel("BMI-Klassen")
plt.ylabel("Durchschnittsalter")
plt.title("Durchschnittliches Alter pro BMI-Klasse")
plt.show()
```
