```python
import pandas as pd


# CSV-Datei von der Festplatte laden
file_path = "data/Obesity_Dataset.xlsx"
df = pd.read_excel(file_path)

# Definiere Spalten, welche mit der Ernährung in Verbindung stehen
eating_habits_columns = [
    'Consumption_of_Fast_Food',
    'Frequency_of_Consuming_Vegetables',
    'Number_of_Main_Meals_Daily',
    'Food_Intake_Between_Meals',
    'Liquid_Intake_Daily'
]

#Berechnen Sie einen ungewichteten Durchschnittswert für diese Spalten, um einen neuen „Eating_Habits_Score“ zu erstellen.
df['Eating_Habits_Score'] = df[eating_habits_columns].mean(axis=1)

# Zeigen Sie die ersten Zeilen an, um die neue Spalte zu überprüfen
df[['Eating_Habits_Score'] + eating_habits_columns].head()

print(df)
```
