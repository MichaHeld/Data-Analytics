```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd

# CSV-Datei von der Festplatte laden
file_path = "data/Obesity_Dataset.xlsx"

df = pd.read_excel(file_path)

# Ersetzen von 1 und 2 in der Spalte 'Sex' durch 'Männlich' und 'Weiblich'
df['Sex'] = df['Sex'].replace({1: 'Male', 2: 'Female'})

# Erstellen der Pivot-Tabelle für die Durchschnittswerte von 'Physical_Exercise' nach 'Sex' und 'Class'
pivot_table = df.pivot_table(
    index='Class',
    columns='Sex',
    values='Physical_Excercise'
)

# Erstellen eines Balkendiagramms für eine bessere Übersicht über die Verteilung
ax = pivot_table.plot(kind='bar', figsize=(10, 6))
plt.title("Physical Exercise by Sex and Class")
plt.xlabel("Class")
plt.ylabel("Average Physical Exercise")
plt.legend(title="Sex")

# Bezeichnungen der x-Achse um 90 Grad drehen
ax.set_xticklabels(ax.get_xticklabels(), rotation=0)

plt.show()
```
