```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd

# Excel-Datei von der Festplatte laden
file_path = "data/Obesity_Dataset.xlsx"

# Excel-Datei laden
data = pd.read_excel(file_path)

# Werte in der Spalte 'Consumption_of_Fast_Food' umbenennen
data['Consumption_of_Fast_Food'] = data['Consumption_of_Fast_Food'].replace({1: "Ja", 2: "Nein"})

# Frequenzzählung erstellen
fast_food_counts = data['Consumption_of_Fast_Food'].value_counts().sort_index()

# Balkendiagramm erstellen
plt.figure(figsize=(10, 6))
sns.barplot(x=fast_food_counts.index, y=fast_food_counts.values, palette="Blues")

# Titel und Achsenbeschriftungen
plt.title("Regelmäßiger Fast-Food-Konsum")
plt.xlabel("Fast-Food-Konsum")
plt.ylabel("Anzahl")
plt.show()
```
