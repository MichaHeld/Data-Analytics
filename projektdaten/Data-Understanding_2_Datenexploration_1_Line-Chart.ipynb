```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd

# CSV-Datei von der Festplatte laden
file_path = "data/Obesity_Dataset.xlsx"

data = pd.read_excel(file_path)

# Kategorien fr krperliche Bewegung definieren
exercise_mapping = {
    1: "0 Tage",
    2: "1-2 Tage",
    3: "3-4 Tage",
    4: "5-6 Tage",
    5: "6+ Tage"
}

# Option 1: Datenwerte direkt durch Kategorien ersetzen
data['Physical_Excercise_Cat'] = data['Physical_Excercise'].map(exercise_mapping)

# Visualisierung
plt.figure(figsize=(10, 6))
sns.lineplot(data=data[:1610], x='Age', y='Physical_Excercise', marker='o', linestyle='-', color='b')

# Achsenticks mit Kategorien beschriften
plt.yticks([1, 2, 3, 4, 5], ["0 Tage", "1-2 Tage", "3-4 Tage", "5-6 Tage", "6+ Tage"])

# Titel und Achsenbeschriftungen
plt.title("Krperliche Bewegung je Alter")
plt.xlabel("Alter")
plt.ylabel("Bewegung pro Woche")

# Diagramm anzeigen
plt.show()
```
