```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
from pandas.api.types import CategoricalDtype

# CSV-Datei von der Festplatte laden
file_path = "data/Obesity_Dataset.xlsx"

# Excel-Datei laden
df = pd.read_excel(file_path)

# Klassen ersetzen
class_mapping = {
    1: 'Untergewicht',
    2: 'Normalgewicht',
    3: 'Übergewicht',
    4: 'Adipositas'
}
df['Class'] = df['Class'].map(class_mapping)  # Klassenwerte durch Kategorien ersetzen

# Reihenfolge der Kategorien definieren
class_order = CategoricalDtype(
    categories=['Adipositas', 'Übergewicht', 'Normalgewicht', 'Untergewicht'],
    ordered=True
)
df['Class'] = df['Class'].astype(class_order)  # Kategorien mit definierter Reihenfolge

# Scatterplot erstellen
plt.figure(figsize=(10, 6))
sns.scatterplot(x='Age', y='Class', data=df)

# Titel und Achsenbeschriftungen
plt.title('Beziehung zwischen Alter und Gewichtsklasse')
plt.xlabel('Alter')
plt.ylabel('Gewichtsklasse')

# Diagramm anzeigen
plt.show()
```
